```python
missing_values_df = missing_values_table(df)
print("Missing Values in df:")
print(missing_values_df)
```

```python
df = df.drop('NumberOfVehiclesInFleet', axis=1)
```

```python
df.shape
```

```python
columns_to_check = ['kilowatts', 'CapitalOutstanding', 'Gender']

# Use the function to remove rows with missing values in the specified columns
df = remove_rows_with_missing_values(df, columns_to_check)

# Display the cleaned DataFrame
print("\nDataFrame after Removing Rows with Missing Values in 'Age' or 'Salary':")
print(df)
```

```python
filling_methods = {


    'CrossBorder': 'mode',
    'Converted': 'mode',
    'Rebuilt': 'mode',
    'WrittenOff': 'mode',
    'NewVehicle': 'mode',
    'Bank': 'mode',
    'AccountType': 'mode',
    'MaritalStatus': 'mode',
   

    'CustomValueEstimate': 'mean',
   

        # Adding multiple columns for median imputation
}

# Apply the fill_missing_values function using the selected methods
df = fill_missing_values(df, filling_methods)

# Display the DataFrame after filling missing values
print("\nDataFrame after Filling Missing Values with Median:")
print(df)
```
